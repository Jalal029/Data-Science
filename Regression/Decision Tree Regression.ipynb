```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor
```

```python
dataset = pd.read_csv('C:\Posisi_gaji.csv')
print(dataset)
X       = dataset.iloc[:,1:2].values
y       = dataset.iloc[:,2].values
```

```python
regressor = DecisionTreeRegressor(random_state = 0)
regressor.fit(X,y)
```

# Memprediksi hasil model dengan input 6.5

```python
y_pred = regressor.predict([[6.5]])
```

# Visualisasi hasil decision tree

```python
X_grid      = np.arange(min(X),max(X), 0.01)
X_grid      = X_grid.reshape((len(X_grid),1))
plt.figure(figsize=(5,50))
plt.scatter(X,y, color = 'salmon')
plt.plot(X_grid, regressor.predict(X_grid), color = 'green')
plt.title('', fontsize = 17, color = 'black')
plt.xlabel('Level posisi',fontsize = 12)
plt.ylabel('Gaji', fontsize=12)
plt.yticks(y,y.astype(int))
plt.show()
```
