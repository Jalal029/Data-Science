```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
dataset = pd.read_csv('C:\Daftar_gaji.csv')
X = dataset.iloc[:,:-1].values
y = dataset.iloc[:, 1].values
```

CREATE TRAIN and TEST SET

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state = 0)
```

FITTING SIMPLE LINEAR REGRESSION terhadap train set

```python
linreg = LinearRegression()
linreg.fit(X_train,y_train)
```

MEMPREDIKSI HASIL TRAINING dan TEST SET

```python
y_pred = linreg.predict(X_test)
y_trainset = linreg.predict(X_train)
```

VISUALISASI HASIL TRAINING SET

```python
plt.scatter(X_train, y_train, color='red')
plt.plot(X_train, y_trainset, color='blue')
plt.title('Gaji vs Pengalaman ( Training Set)',color='black')
plt.xlabel('Tahun bekerja',fontsize = 13)
plt.ylabel('Gaji',fontsize=13)
plt.show()
```

VISUALISASI HASIL TEST SET

```python
plt.scatter(X_test, y_test, color='green')
plt.plot(X_train, y_trainset, color='yellow')
plt.title('Gaji vs Pengalaman',fontsize=18,color='red')
plt.xlabel('Gaji',fontsize=14)
plt.ylabel('Lama bekerja',fontsize=14)
plt.show()
```
