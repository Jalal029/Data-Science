```python
movie_df['startYear'] = movie_df['startYear'].replace('\\N', np.nan)
movie_df['startYear'] = movie_df['startYear'].astype(float)

movie_df['endYear'] = movie_df['endYear'].replace('\\N', np.nan)
movie_df['endYear'] = movie_df['endYear'].astype(float)

movie_df['runtimeMinutes'] = movie_df['runtimeMinutes'].replace('\\N', np.nan)
movie_df['runtimeMinutes'] = movie_df['runtimeMinutes'].astype(float)

movie_df.isna().sum()
```

```python
# mengubah nilai genres ke list
def transform_to_list(x):
    if '\\N' in x: 
        return []
    else: 
        return x.split(',')

movie_df['genres'] = movie_df['genres'].apply(lambda x: transform_to_list(x))
```

# Weighted Formula

```python
def imdb_weighted_rating(df, var=0.7):
    v = df['numVotes']
    m = df['numVotes'].quantile(var)
    R = df['averageRating']
    C = df['averageRating'].mean()
    df['score'] = (v/(v+m))*R + m/(m+v)*C
    return df['score']

imdb_weighted_rating(movie_rating_df)
```

# simple recommender system ( TOP 10 )

```python
def recommender(df, top):
    df = df.loc[df['numVotes'] >= m]
    df = df.sort_values('score',ascending=False)
    
    df = df[:top]
    return df
    
recommender(movie_rating_df, 11)
```

# recommender system with user preferences

```python
df = movie_rating_df.copy()

def user_prefer_recommend(df, adult, startyear, genre, top):
    #Apakah dewasa atau tidak
    if adult == 'Yes':
        df = df.loc[df['isAdult'] == 1]
    elif adult == 'No':
        df = df.loc[df['isAdult'] == 0]
    
    df = df.loc[df['startYear'] >= startyear]
    
    if genre == 'All':
        df = df
    
    else :
        def filter_genre(x):
            if genres == str(x):
                return True
            else :
                return False
        df = df.loc[df['genres'].apply(lambda x:filter_genre(x))]
        
    df = df.loc[df['numVotes'] > m]
    df = df.sort_values(by='score', ascending=False)
    
    df = df[:top]
    return df
```

```python
user_prefer_recommend(movie_rating_df, 'No', 2012, 'All', 11)
```
