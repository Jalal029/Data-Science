```python
import pandas as pd
```

```python
housing_df = pd.read_csv('C:/Dataset/iowa_data.csv')
housing_df.head()
```

```python
housing_df.info()
```

# SalePrice sebagai target

```python
housing_df.dropna(axis = 0, subset=['SalePrice'], inplace=True)
```

```python
housing_df.shape
```

```python
# tidak ada missing values
```

```python
y = housing_df['SalePrice']
y.head()
```

```python
X = housing_df.drop('SalePrice', axis=1)
X = X.select_dtypes(exclude=['object'])
X.head()
```

# training dan tes set

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
X_train.head()
```

# Model

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
rf_model = RandomForestRegressor(random_state = 0)
```

Fungsi untuk mengukur performa dari dataset model dengan MAE

```python
from sklearn.metrics import mean_absolute_error
```

```python
def score_dataset(X_train, X_test, y_train, y_test, model=rf_model):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    return mean_absolute_error(y_test, y_pred)
```

Identifikasi missing values

```python
X_train.shape
```

```python
na_cols = X_train.isna().sum()
na_cols[na_cols>0]
```

```python
sum(na_cols[na_cols>0])
```

# Penanganan terhadap missing values

1. Drop kolom yang mengandung missing values

```python
cols_with_na = [col for col in X_train.columns if X_train[col].isna().any()]
cols_with_na
```

Me-remove columns missing values

```python
X_train_reduced = X_train.drop(cols_with_na, axis=1)
X_test_reduced  = X_test.drop(cols_with_na, axis=1)
```

```python
print(f'X_train : {X_train_reduced.shape}, X_test : {X_test_reduced.shape}')
```

Mengukur performa dataset setelah di reduced columns

```python
score_dataset(X_train_reduced, X_test_reduced, y_train, y_test)
```

2. Imputation Missing Values

```python
from sklearn.impute import SimpleImputer
```

```python
my_imputer = SimpleImputer()
my_imputer.fit(X_train)
```

```python
X_train_imputed = pd.DataFrame(my_imputer.transform(X_train))
X_test_imputed  = pd.DataFrame(my_imputer.transform(X_test))
X_train_imputed.head()
```

```python
X_train_imputed.columns = X_train.columns
X_train_imputed.head()
```

```python
score_dataset(X_train_imputed, X_test_imputed, y_train, y_test)
```
