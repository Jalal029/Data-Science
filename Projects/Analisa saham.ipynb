```python
import pandas as pd
import matplotlib.pyplot as plt
import math
import pandas_datareader.data as web
import datetime
import seaborn as sns
```

```python
start = datetime.datetime(2020,1,1)
end   = datetime.datetime(2021,2,2)

df    = web.DataReader('INDF.JK','yahoo', start=start, end=end)
```

```python
df.tail()
```

```python
len(df)
```

```python
close_px = df['Adj Close']
mavg     = close_px.rolling(window=30, center=True).mean()
```

```python
plt.plot(close_px)
plt.plot(mavg)
plt.show()
```

```python
rets = close_px / close_px.shift(1) - 1
rets.plot()
```
