# Methode Mapping

1. method map ( diterapkan pada series )

```python
points_mean = wine_df['points'].mean()
wine_df['points'].map(lambda x: x - points_mean)
```

2. method apply ( diterapkan di DataFrame)

```python
def remean_points(row):
    row['points'] = row['points'] -  points_mean
    return row

wine_df.apply(remean_points, axis=1).head() 
```

Kasus : mencari jumlah wine dengan description tropical dan fruity. Siapkan series untuk menampung jumlah data

```python
n_tropical = wine_df['description'].map(lambda x: 'tropical' in x).sum()
n_fruity   = wine_df['description'].map(lambda x: 'fruity' in x).sum()
pd.Series([n_tropical, n_fruity], index=['tropical','fruity'])
```

# melakukan konversi point menjadi star

```python
def converter(row):
    if row['country'] == 'US':
        row['star'] = 3
    elif row['country'] == 'Italy':
        row['star'] = 3.5
    elif row['points'] >= 92:
        row['star'] = 2.8
    elif row['points'] >= 85:
        row['star'] = 2.5
    else :
        row['star'] = 1
    return row

wine_df['star'] = None
wine_df.apply(converter, axis=1).head()
```
